# Validación cruzada

Es una técnica utilizada para evaluar el desempeño de los resultados de un análisis y garantizar que son independientes de la preatición entre datos de entrenamiento y prueba.

## Procedimiento

1. Se divide la data en $k$ grupos de tamaño similar.
2. Se usan $k-1$ grupos para entrenar un modelo y el excluido, para validarlo.
3. El proceso se repite $k$ veces, usando un grupo de validación distinto en cada iteración.


```python
import pandas as pd
import numpy as np
import io
```

```python
data = pd.read_csv('data/cancerdata.csv', index_col=0)
data.head()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=4)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
score
```

```python
from sklearn.model_selection import KFold
```

```python
model = DecisionTreeClassifier()
kfold_validation = KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
resultados = cross_val_score(model, X, y, cv = kfold_validation)
resultados.mean()
```
